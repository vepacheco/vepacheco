```python
for i in range(len(df['abilities'])):
    df['abilities'][i] = df['abilities'][i].replace('[','')
    df['abilities'][i] = df['abilities'][i].replace(']','')
    df['abilities'][i] = df['abilities'][i].split(',')
```

```python
from sklearn.preprocessing import LabelBinarizer
label_binarizer = LabelBinarizer()
label_binarizer_output = label_binarizer.fit_transform( df['abilities'])
```
